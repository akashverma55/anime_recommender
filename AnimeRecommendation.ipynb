```python
import ast
def convert3(obj):
  L=[]
  counter=0
  for i in ast.literal_eval(obj):
    if counter !=3:
      L.append(i['name'])
      counter+=1
    else:
      break
  return L
# anime_daat['cast']=anime_daat['cast'].apply(convert3)
```

```python
def convert(obj):
  L=[]
  for i in ast.literal_eval(obj):
    L.append(i['name'])
  return L
# anime_daat['genre']=anime_daat['genre'].apply(convert)
```

```python
def fetch_director(obj):
  L=[]
  for i in ast.literal_eval(obj):
    if i['job']=='Director':
      L.append(i['name'])
      break
  return L
# anime_daat['director']=anime_daat['director'].apply(fetch_director)
```

```python
def join_producers(producer_list):
  return ''.join(producer.replace(" ", "") for producer in producer_list)
```

```python
anime_data['Producers'] = anime_data['Producers'].apply(join_producers)
```

```python
anime_data.head()
```

```python
anime_data['Tags']=anime_data['Genres']+anime_data['Type']+anime_data['Producers']+anime_data['Studios']
```

```python
anime_data.head()
```

```python
new_anime_data=anime_data[['MAL_ID','Name','Tags']]
```

```python
new_anime_data.head()
```

```python
def join_with_space(producer_list):
  return ''.join(producer.replace(",", " ") for producer in producer_list)
```

```python
new_anime_data['Tags']=new_anime_data['Tags'].apply(join_with_space)
```

```python
new_anime_data.head()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
  y=[]
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
new_anime_data['Tags']=new_anime_data['Tags'].apply(stem)
```

```python
new_anime_data['Tags'][0]
```

```python
new_anime_data['Tags'] = new_anime_data['Tags'].apply(lambda x:x.lower())
```

```python
anime_data.shape
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=1000,stop_words='english')
```

```python
word_vectors=cv.fit_transform(new_anime_data['Tags']).toarray()
```

```python
word_vectors[0]
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
theta=cosine_similarity(word_vectors)
```

```python
theta.shape
```

```python
theta[0]
```

```python
def recommend(anime):
  anime_index = new_anime_data[new_anime_data["Name"] == anime].index[0]
  distances = theta[anime_index]
  anime_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:11]
  for i in anime_list:
    print(new_anime_data.iloc[i[0]]["Name"])
```

```python
new_anime_data['Name'] = new_anime_data['Name'].apply(lambda x:x.lower())
```

```python
new_anime_data.tail()
```

```python
recommend("mieruko-chan")
```

```python
import pickle
pickle.dump(new_anime_data.to_dict(),open('anime.pkl','wb'))
```
